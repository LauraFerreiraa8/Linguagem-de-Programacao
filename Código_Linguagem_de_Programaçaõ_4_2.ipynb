##3. Desenvolvimento

###3.1. Parte 1: Variáveis e Saída de Dados "Pythonic"
Em Python, tudo é um objeto. A tipagem dinâmica dispensa a declaração prévia de tipos, e as F-Strings (PEP 498) facilitam a exibição de dados.

```python
# Definindo variáveis (Tipagem Dinâmica)
nome = "Analista de Sistemas"  # str
idade = 25                      # int
altura = 1.75                   # float
ativo = True                    # bool

# Saída formatada com F-Strings
print(f"Perfil: {nome}")
print(f"Idade: {idade} anos | Altura: {altura}m")
print(f"Status do Perfil: {'Ativo' if ativo else 'Inativo'}")

# Verificando a natureza do objeto
print(f"A variável 'altura' é da classe: {type(altura)}")
```

####O que observar: O interpretador define o tipo na atribuição. O uso de chaves {} dentro de uma string precedida por f permite injetar lógica e variáveis diretamente no texto, mantendo o código limpo.



###3.2. Parte 2: Operações e Precedência Matemática
Além das operações básicas, Python oferece operadores para divisão inteira (//), resto (%) e exponenciação ().

```python
# Simulação de cálculos matemáticos
valor_a = 15
valor_b = 4

# Operadores específicos
div_inteira = valor_a // valor_b
resto_div = valor_a % valor_b
potencia = valor_a ** 2

# Teste de Precedência: () > ** > * / > + -
resultado_complexo = (10 + 5) * 2 ** 2

print(f"Divisão Inteira de 15 por 4: {div_inteira}")
print(f"Resto da divisão: {resto_div}")
print(f"Resultado da expressão (10+5)*2^2: {resultado_complexo}")
```

###3.3. Parte 3: Entrada de Dados e Tratamento de Erros (Casting)
Um erro comum (TypeError) ocorre ao tentar operar dados vindos de input(), pois esta função sempre retorna strings. O "Casting" é a solução técnica.

```python
# Simulação de Cálculo de Equação / Área
try:
    print("--- Calculadora de Área ---")
    base_str = input("Digite o valor da base: ")
    altura_str = input("Digite o valor da altura: ")

    # Casting: convertendo texto para número decimal
    base = float(base_str)
    altura = float(altura_str)

    area = (base * altura) / 2
    print(f"A área do triângulo é: {area:.2f}")

except ValueError:
    print("ERRO: Certifique-se de digitar apenas números!")
```

####O que observar: O bloco try/except captura falhas de conversão. Se o usuário digitar uma letra, o programa não interrompe bruscamente, mas exibe uma mensagem de orientação, garantindo a robustez do software.
